# **Updating Variables**

```python
x=5
print(x)
```

# **Hard Coding**

Hard coding means writing down the answer without actually going through the work of computing the answer. Suppose you want to calculate the average of the two numbers with the help of following code:

```python
x=20
y=30

avg=25.0
print(avg)
```

```python
x=10
y=20

avg=25.0
print(avg)
```

We can see in the above code that we have done hard coding to get average but we are not getting correct output. Instead we can write an expression for computing the average.

```python
x=10
y=20

avg=(x+y)/2
print(avg)
```

# **Input**

Input function asks the user to enter something which accepts the prompt as an argument. Let's understand with the help of following code:

```python
n=input("Please Enter Your Name: ")
print("Hello",n)
```

Input function always returns the value in 'string' type. So in oder to use them as a numeric type we have to cast them. Look below example:

```python
str_seconds=input("Please enter the number of seconds you wish to convert: ")
total_secs=int(str_seconds)  #Casting the string to numeric type

hours=total_secs//3600
secs_still_remaining=total_secs%3600
mins=secs_still_remaining//60
secs_finally_remaining=secs_still_remaining%60
print("Hrs=",hours, "minutes=",mins, "secs=",secs_finally_remaining)
```
