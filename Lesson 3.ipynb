# **Variables**
Variables allow us to store the value of the given expression. Let's understand by below examples:

```python
message="Name is Sherlock Holmes, Address is 221b Baker Street"
a=19
pi=3.14159
```

The above used variables are called 'Setting Variables'. 

```python
print(message)
print(a)
print(pi)
```

The above variables are called 'Referencing Variable'. But what happens when we override the value of the variable. 

```python
x=7
print(x)

x=8
print(x)
```

Every variable name should start with letters and should only contain letters and numbers. Python is case sensitive.

```python
XYZ=100
xyz=200
print(XYZ)
print(xyz)
```

Python treats '_' as a character and can be often used as a space while giving name to the variables.

```python
my_variable=30
print(my_variable)
```

# **Statements and Expressions**
We know that expression is a building block of the program. The statement can be defined as following example:

x=1+2+3

where, 'x=1+2+3' is called statement while '1+2+3' is called an expression. 

Python does the work of computing the value of the expression.  

```python
a=3.14
b=len("Bazinga!")
print(a)
print(b)
```

We can combine expressions in terms of larger expression as well.

```python
print(2*len("Oh Leonard")+len("You magnificent beast "))
```
