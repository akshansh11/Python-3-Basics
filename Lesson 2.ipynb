#**Function**

Function can be thought of as some machinery which takes in some inputs (arguments), process them and spits out the output (return value). Now let's see how it can be expressed in terms of Python programming. For example, we have a function named 'square' and it takes 5 as an argument. It should be noted that the function can take any number of arguments but has only one return value. 

square(5): In this case square is a function and 5 is an argument.

print(square(5)): In this case print is the function and 'square(5)' is an argument.

print(sub(6,4)): This function will return the output as 2. 

print(square(5)+sub(6,4)): Function starts to process from inside arguments to outside. First it will process 'sub(6,4)' then it moves towards 'square(5)' argument.
